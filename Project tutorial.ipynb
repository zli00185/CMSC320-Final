Next, we will randomly try some columns and plot them to see what kind of relation we can find. 

```python
# Comparing streaming time v/s Followers
plt.figure(figsize=(12,8))
sns.lineplot(streamer['Stream time(minutes)'], streamer['Followers']);
plt.title('Streaming time v/s Followers', fontsize = 20);
plt.show();
```

```python
# Comparing streaming time v/s Average viewers
plt.figure(figsize=(12,8)) #comparing streaming time v/s average viewers
sns.lineplot(streamer['Stream time(minutes)'], streamer['Average viewers'], palette = "Set1")
plt.title('Streaming time v/s Average Viewers', fontsize = 20)
plt.show()
```

From these two figures, it can be seen that the Stream time is inversely proportional to the average number of viewers and followers gained. Especially for the data on the number of Average viewers, as the stream time increase, it decreases significantly in the first 50000 minutes. I think the main reason is that the streamer's style and the type of viewers he or she would attract have been gradually determined in those amount of streaming time, so most of the unattracted audience will leave, and the others loyal audience will stay. This logic might also apply to followers. Followers are the most loyal group of viewers. Their payment, tipping habits, and the amount will determine the streamer's income and popularity. Although the graph of stream time vs followers is less regular than that of average views, we can still find that most followers have been finalized in the early and mid-term of streaming time. But there is a possibility that there isn't any relation between followers and stream time, and we will test it later.

```python
# Comparing streaming time v/s watch time
plt.figure(figsize=(12,8)) #comparing streaming time v/s Watch time(Minutes)
sns.lineplot(streamer['Stream time(minutes)'], streamer['Watch time(Minutes)']);
plt.title('Streaming time v/s Watch time(Minutes)', fontsize = 20);
plt.show();
```

Another interesting thing I found is that longer streaming time does not represent more watch time from viewers. The table above proves this point clearly. The highest peak is the data from streamer xqc, who has the most watch time from viewers. However, his streaming time is only half of the one with the most stream time (RiffTrax). The case of xqc is not accidental. Summit1g, who ranks second in the watch time list, even has less streaming time than xqc. Only the third-ranked Gaules has the high watch time with also high streaming time, while other streamers whose streaming time is as long as Gaules have much lower watch time value.

```python
# comparing Average viewers v/s Followers
plt.figure(figsize=(12,8)) 
sns.scatterplot(streamer['Average viewers'], streamer['Followers']);
sns.regplot(x = 'Average viewers', y = 'Followers', data = streamer);
plt.title('Average viewers v/s Followers', fontsize = 20);
plt.show();
```

```python
# Print the summary of linear regression
x = streamer['Average viewers']
y = streamer['Followers']
result = sm.OLS(y, x).fit()
print(result.summary())
```

From the scatterplot and positive regression line above, we can conclude that Average viewers and Followers are positively related. This indicates that more average views usually refer to more followers. According to the summary of the linear regression line, we can say one average viewer increase could lead to 59.7879 more followers.

However, we could find two exceptions with the two highest average viewer counts and low follower counts at the bottom right of the scatterplot. Those two channels are dota2ti and dota2ti_ru, which are two official streaming accounts for the video game Dota2. These two accounts only stream when there is an official gaming championship for Dota2. Therefore, in terms of operation strategy, they are not streaming as frequently as other individual streamers, and those large numbers of average viewers will not become followers because most of them are here to watch matches.

```python
# Plot the head map of correlation between each column
plt.figure(figsize=(12,8))
sns.heatmap(streamer[['Channel', 'Watch time(Minutes)', 'Stream time(minutes)', 'Followers','Peak viewers','Average viewers','Followers gained','Views gained','Partnered','Mature','Language']].corr(), annot = True) #overall correlation between the various columns present in our data
plt.title('Overall relation between columns of the Dataset', fontsize = 20)
plt.show()
```

The heat map shows the overall relation between columns. We can see that Watching time has a middle relation with column followers, peak views, the follower gained, views gained due to R > 0.5, and weak relation with average viewers due to R > 0.25. Column Followers has a strong relation with column followed gained due to R > 0.7, middle relation with column watch time, peak viewers due to R >0.5, and weak relation with column Average viewers, views gained due to R > 0.25.

From the chart, we get to know that the previous graphs of stream time vs. follower and stream time vs. watch time were coming from columns with relatively low correlation values (-0.091 and 0.15), so we couldn't find too much information there. The correlation value of Stream time vs. average viewers (-0.25) is relatively weak, but it indicates the hypothesis we made early was correct. Meanwhile, the -0.091 correlation also shows that there isn't any relation between followers and stream time, which surprises me since most of us will think people with more followers tend to stream more, and people who stream more get more followers. However, it seems like that is not how it works.

We will look at the relation between columns with the highest correlation value next, which are follower vs follower gained (0.72).

```python
# Comparing followers v/s followers gained
plt.figure(figsize=(12,8)) #comparing streaming time v/s Watch time(Minutes)
sns.scatterplot(streamer['Followers'], streamer['Followers gained']);
sns.regplot(x = 'Followers', y = 'Followers gained', data = streamer);
plt.title('Followers v/s Followers gained', fontsize = 20);
plt.show();
```

```python
# Print summary of linear regression
x = streamer['Followers']
y = streamer['Followers gained']
result = sm.OLS(y, x).fit()
print(result.summary())
```

According to the summary of line, coef of 0.3218 implys every 3 followers will refers to 1 gained follower. This represents follower gained are increasing as the total size followers increase. A P-value of 0 also indicate we reject the H0.
